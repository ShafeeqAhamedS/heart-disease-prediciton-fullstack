```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./Heart_Disease_Prediction.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Heart Disease'] = df['Heart Disease'].replace('Presence', 1)
df['Heart Disease'] = df['Heart Disease'].replace('Absence', 0)
df
```

```python
# Normalizing the data
scaler = StandardScaler()
scaler.fit(df.drop('Heart Disease', axis=1))
scaled_features = scaler.transform(df.drop('Heart Disease', axis=1))
df_feat = pd.DataFrame(scaled_features, columns=df.columns[:-1])
df_feat
```

```python
df
```

```python
# Encode target variable
target = 'Heart Disease'
X = df_feat
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential([
    Dense(32, input_dim=X_train.shape[1], activation='relu'),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dropout(0.2),
    Dense(8, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Define early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# Train the model
history = model.fit(X_train, y_train,
                    validation_split=0.2,
                    epochs=100,
                    batch_size=16,
                    verbose=1)
```

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

print(f"Loss: {loss}, Accuracy: {accuracy}")
```
